```python
import matplotlib.pyplot as plt
import re

from typing import List
```

```python
def plot(y: List[int]) -> None:
    x = [0.01 * i for i in range(1, len(y) + 1)]

    plt.plot(x, y)

    plt.xlabel('time(seconds)')
    plt.ylabel('Error Rate')
    plt.title('Error rate by time')

    plt.show()

def get_timings(file_path: str) -> List[int] | None:
    lst = None
    with open(file_path, 'r', encoding='UTF-8') as file:
        lst = file.readline()

    if lst:
        return [int(val) for val in re.sub(r'[\[\],]', '', lst).split()]
    return None
```

```python
print(get_timings("files.txt"))
```

```python
plot(get_timings("files_idle.txt"))
```
